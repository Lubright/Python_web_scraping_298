```python
# coding=UTF-8
import requests
from pyquery import PyQuery as pq
```

```python
html_content=requests.get('https://www.webscraper.io/test-sites/e-commerce/allinone')
html_content.encoding = 'utf-8' # default is ISO-8859-1
print("encoding: %s" % html_content.encoding)
print("content: \n%s" % html_content.text)
print('type of html_content:',type(html_content.text))
```

```python
doc=pq(html_content.text)
doc
```

```python
h1=doc('div.jumbotron > h1')
# print(h1)
print(h1.text())
```

```python
description_text=doc('div.jumbotron > p')
# print(description_text)
print(description_text.text())
```
